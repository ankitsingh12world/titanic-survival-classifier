### IQR Method

```python
dfship["Age"]
```

```python
q1 = dfship["Age"].quantile(0.25)
print(q1)
```

```python
q3 = dfship["Age"].quantile(0.75)
print(q3)
```

```python
iqr = q3 - q1
```

```python
lowerlimit = q1 - (1.5 * iqr)
```

```python
upperlimit = q3 + (1.5 * iqr)
```

### Cross Validation

```python
from sklearn.model_selection import cross_val_score
```

```python
knn1 = KNeighborsClassifier()
```

### Cross Validation with K-Fold

```python
from sklearn.model_selection import cross_val_score, KFold, StratifiedKFold
```

```python
knn2 = KNeighborsClassifier()
```

```python
knn2
```

```python
kf = KFold(n_splits=20)
```

```python
score2 = cross_val_score(knn2, X_ros, y_ros, cv = kf, n_jobs = -1)
print(score2)
```

```python
score2.mean()
```
